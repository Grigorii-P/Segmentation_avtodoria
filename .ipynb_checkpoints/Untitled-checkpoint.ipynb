```python
import numpy as np
import matplotlib.pyplot as plt

def a_b(c1, c2):
    x1, x2, y1, y2 = c1[0], c2[0], c1[1], c2[1]
    b = (y2 - y1 * x2**2 / x1**2) / (x2 - x2**2 / x1)
    a = (y1 - b * x1) / x1**2
    return a, b
    
    
c1 = (3071,191)
c2 = (4095, 255)

a, b = a_b(c1, c2)   

def f(t):
    return a * t**2 + b * t
```

```python
r1, r2 = -800000, 800096
t1 = np.arange(r1, r2, 50)
plt.figure(1, figsize=(12,7))
plt.plot(t1, f(t1), 'r--')
plt.show()
```
